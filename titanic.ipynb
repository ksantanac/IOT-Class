```python
# Localiza todos as linhas cujos valores não são numericos na coluna classe_social e impra a tupla que mostra a dimensao
# .loc -> Quais linhas analisar e qual coluna
# .shape -> somar e agrupar em tupla
# isna -> retorna número da linha

df.loc[pd.isna(df['sexo']), 'sexo'].shape

# df_coluna = df.dropna(subset=['classe_social'])

```

```python
indefinidos_idade = df.loc[pd.isna(df['idade']), 'idade'].shape[0] #pega o valor 177
total = df.idade.shape[0]
print("Total: ", total)
print((indefinidos_idade/total) * 100) 
```

```python
def calcula_percentual_indefinido(df, coluna):
    indefinidos = df.loc[pd.isna(df[coluna]), coluna].shape[0]
    total = df[coluna].shape[0]

    return indefinidos/total * 100
```

```python
df = df.drop(columns=["cabine"])
df.head(50)
```

```python
# Quantidade  de linhas faltantes na coluna. insa-> todas as linhas que são nans e soma
df.isna().sum()
```

```python
# deleta todas a linhas cujos números não são valores válidos | isfinite -> verfica todas as linhas que tem números finitos, ou seja, != NAN e ignora.
df = df[np.isfinite(df['idade'])]
df.head(50)
```

```python
df = df[~pd.isna(df.embarque)] # todos números != de NAN
calcula_percentual_indefinido(df, 'embarque')
```
